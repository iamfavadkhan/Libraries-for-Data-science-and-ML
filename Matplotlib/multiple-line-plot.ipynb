```python
import matplotlib.pyplot as plt 
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']

# Temperature of City 1 in °C
city1 = [30, 32, 33, 29, 35]

# Temperature of City 2 in °C
city2 = [25, 27, 26, 30, 28]

plt.plot(days , city1 , label = "city 1" , color = "red" , linestyle='-', marker='o')
plt.plot(days , city2 , label = "city 2" , color = "green" , linestyle='--', marker='s')
plt.legend()
plt.grid()
plt.title("Temperature Comparision between city1 and city2")
plt.xlabel("Temperature in °C")
plt.ylabel("Days")
plt.show()
```
