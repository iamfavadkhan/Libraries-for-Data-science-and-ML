```python
import matplotlib.pyplot as plt 
products = ['A', 'B', 'C', 'D']
sales = [120, 300, 250, 400]

colors = ['r','g','b']
plt.bar(products , sales , color = colors)
plt.xlabel("Products" , size = 16)
plt.ylabel("Sales" , size = 16)

for i in range(len(products)):
    plt.text(products[i], sales[i] + 3, f"({products[i]}, {sales[i]})", ha='center', color='blue')

plt.show()
```
