```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('weather.csv')
print(df.head())

max_temp = df['MaxTemp']
min_temp = df['MinTemp']
rainy_days = df[df['Condition'] == 'Rainy']

days = df['Day']


plt.figure(figsize=(12,6))
plt.plot(days , max_temp , color = 'red' , marker = 'o' , label = 'Maximum Temperature' , linestyle = '-')
plt.plot(days , min_temp , color = 'Blue' , marker = 's' , label = 'Minimum Temperature' , linestyle = '--') 
plt.scatter(rainy_days['Day'], rainy_days['MaxTemp'] , color='green', s=100, label='Rainy Days', zorder=5)
plt.scatter(rainy_days['Day'], rainy_days['MinTemp'], color='green', s=100, label='Rainy Days (Min)', zorder=5, marker='D')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Temperature in °C")
plt.title('30 Days Temperature Analysis of Lahore')
plt.grid('True')
plt.show()
```
