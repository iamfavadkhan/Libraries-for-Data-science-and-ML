```python
import matplotlib.pyplot as plt 

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperature = [7, 9, 14, 18, 23, 27, 30, 29, 25, 19, 13, 8]   # Â°C
rainfall = [78, 60, 55, 48, 30, 20, 10, 12, 18, 35, 58, 72]  # mm
humidity = [85, 80, 75, 70, 65, 60, 55, 58, 62, 70, 78, 82]  # %

#plot 1
plt.subplot(2,2,1)
plt.plot(temperature , months)
plt.xlabel("Months")
plt.ylabel("Temperature Â°C")

#plot 2
plt.subplot(2,2,2)
plt.bar(rainfall , months)
plt.xlabel("Rain Fall")
plt.ylabel("Months")

#plot 3
plt.subplot(2,2,3)
plt.plot(humidity , months)
plt.xlabel("Humadity")
plt.ylabel("Months")

#plot 4
plt.subplot(2, 2, 4)
plt.plot(temperature ,months , label="Temperature", marker='o')
plt.plot(rainfall, months ,label="Rainfall", marker='s')
plt.plot(humidity, months ,label="Humidity", marker='^')
plt.xlabel("Months")
plt.ylabel("Values")
plt.title("Weather Comparison")
plt.legend()
plt.tight_layout()
plt.show()
```
