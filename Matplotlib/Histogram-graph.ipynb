#Drawing a histogram graph

```python
import random 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

numbers = np.random.randint(10 , 250 , size = 150)
print(numbers)
```

```python
colors = ['red']
plt.hist(numbers,color=colors,edgecolor="blue",label="python")
plt.legend
plt.show
```
