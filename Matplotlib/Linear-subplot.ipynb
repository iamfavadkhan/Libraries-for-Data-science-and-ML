```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

x = np.array([2,3,4,6,7,9])
y = np.array([0,3,3,5,6,7])

plt.subplot(2,1,2)
plt.plot(x , y)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

plt.subplot(2,2,1)   # 1 row, 2 columns, plot 1
plt.plot(x, y1)
plt.title("Square")

plt.subplot(2, 2, 2)   # 1 row, 2 columns, plot 2
plt.plot(x, y2)
plt.title("Linear")

plt.tight_layout()
plt.show()
```
