#Creating Bar Graph and Applying It's basic methods

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

x = np.array(["Python" , "CPP" , "Java" , "C#"])
y = np.array([85,60,75,50])

plt.title("Popularity Level of Programming Languages")
plt.xlabel("Programming Languages")
plt.ylabel("Percentage Level")
c = ['r','g','b']
plt.bar(x,y , width=0.5 , color = c)
# plt.legend()
for i in range(len(x)):
    plt.text(x[i] , y[i]+1 , f"{x[i]} , {y[i]}" , ha = 'center' , color = 'blue')
plt.show()
```

#Bar in the horizantal line

```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.array(["Python" , "CPP" , "Java" , "C#"])
y = np.array([85,60,75,50])
plt.barh(x, y)
plt.show()
```

#Colors  and Bar width

```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.array(["Python" , "CPP" , "Java" , "C#"])
y = np.array([85,60,75,50])
colors = ["red" , "blue" , "green" , "black"]
plt.bar(x, y , color = colors , width=0.5)
plt.show()
```

#Bar height

```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.array(["Python" , "CPP" , "Java" , "C#"])
y = np.array([85,60,75,50])
plt.bar(x, y , height = 0.1)
plt.show()
```
