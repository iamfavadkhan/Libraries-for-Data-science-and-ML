#Scatter graph and som operation


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

x = [2,3,4,5,5,6,7,8,17]
y = [2,4,5,4,6,6,7,9,2,]
plt.scatter(x , y , color = "#79b98c")

x = [12,2,4,8,9,11,18]
y = [21,14,15,4,16,11,27]
colors = np.array(["red","green","blue","yellow","pink","black","orange"])
plt.scatter(x , y , color = colors)

plt.title("Scatter Plot" , size = 16)
plt.xlabel("Number of Rows")
plt.ylabel("Number of columns")


plt.show()
```

#Combine color and size and aplha

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
