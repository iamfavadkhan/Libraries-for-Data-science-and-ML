#Drawing a linear Graph

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

x = np.array([1,2,3,4,7])
y = np.array([2,4,6,7,4])

c = ('r','g','b')
plt.plot(x , y )
plt.title("Simple Linear Graph")
plt.xlabel("No.")
plt.ylabel("Count.")
plt.show()
```

#Plotting without Lines

```python
plt.plot(x , y , 'o')
plt.show()
```

#Default x points

```python
y = np.array([2,4,6,7,4])

plt.plot(y)
plt.show()
```

#Markers

```python
plt.plot(x , y , marker = "*")
plt.show()
```

#Line Plotting

```python
plt.plot(x,y, 'o:r')
plt.show()
```

```python
plt.plot(x,y,  marker = 'o' , ms = 10)
plt.show()
```

#Marker edge color , marker size and marker face color

```python
plt.plot(x,y,  marker = 'o' , ms = 10 , mec = "green" , mfc = "red")
plt.show()
```

#For plot line style you can used simple keywords like linestyles or ls

```python
plt.plot(x , y  , ls = 'dotted' , marker = 'o')
plt.show
```

```python
plt.plot(x , y  , ls = 'dashed' , marker = 'o')
plt.show()
```

```python
plt.plot(x , y  , ls = 'dashdot' , marker = 'o')
plt.show
```

#Line Width

```python
plt.plot(x , y  , linewidth = 10 , marker = 'o' , ms = 30)
plt.show
```

#Multiple Lines 

```python
x1 = np.array([2,3,5,6,7,7,8])
y1 = np.array([6,3,5,6,7,5,9])
plt.plot(x , y , x1 , y1 , )
plt.show()
```

#Title , xlabel and ylabel

```python
plt.plot(x , y)
plt.xlabel("Data number ")
plt.ylabel("Data sets ")
plt.title("Data Sets Details")
plt.show()
```

#More on title , xlabel and ylabel

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.plot(x , y)
plt.xlabel("Data number " , fontdict=font1)
plt.ylabel("Data sets " , fontdict=font2)
plt.title("Data Sets Details" , fontdict=font2)
plt.show()
```

#Loc methods

```python
plt.plot(x , y)
plt.xlabel("Data number ")
plt.ylabel("Data sets ")
plt.title("Data Sets Details" , loc = "left")
plt.show()
```

#Adding grid lines to the graph

```python
plt.plot(x , y)
plt.xlabel("Data number ")
plt.ylabel("Data sets ")
plt.title("Data Sets Details")
plt.grid()
plt.show()
```

```python
plt.plot(x , y)
plt.xlabel("Data number ")
plt.ylabel("Data sets ")
plt.title("Data Sets Details")
plt.grid(axis="x")
plt.show()
```

```python
plt.plot(x , y)
plt.xlabel("Data number ")
plt.ylabel("Data sets ")
plt.title("Data Sets Details")
plt.grid(axis="y")
plt.show()
```
