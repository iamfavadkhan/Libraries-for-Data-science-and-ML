#Creating simple Line plot

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

x_1 = [1,2,3,4,5,6,7,5,4]
x_2 = [4,3,2,2,3,4,5,6,7]

x = pd.DataFrame({"var" : x_1, "var2" : x_2})
sns.lineplot(x=x_1 , y=x_2 , data=x)
plt.grid()
plt.show()

```

```python
y_1 = sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm" , y="bill_depth_mm" , data=y_1)
plt.show()
```

hue , size , marker , dash and style parameters

```python
sns.lineplot(x="bill_length_mm" , y="bill_depth_mm" , data=y_1 , hue="sex" , size = 50 , style="sex" , markers=["o" , ">"])
plt.grid()
plt.title("Pyhton Library")
plt.show()
```
