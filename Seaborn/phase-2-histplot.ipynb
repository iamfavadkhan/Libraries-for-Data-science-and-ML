#Histogram

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

data = sns.load_dataset("penguins")
data.fillna('0')
data.head(20)
```

```python
sns.displot(data["bill_length_mm"] )
plt.show()
```

#Bins , kde and rug parameter

```python
sns.displot(data["bill_length_mm"] , bins = [30,40,50,60,70], kde = True , rug=True)
plt.show()
```

#log_scale parameter

```python
sns.displot(data["bill_length_mm"] , log_scale=True , kde = True)
plt.show()
```
