```python
# Task 2: Modify Data in Memory and Write Back
# Read CSV data from a string using StringIO, 
# add a new column calculated from existing ones
# (e.g., Mortality Rate = Deaths/Confirmed * 100),
# and then write the updated DataFrame back to a 
# new CSV string stored in another StringIO object.
# Print the resulting CSV string.
import pandas as pd
from io import StringIO 

covid_data = """Country,Deaths,Confirmed Cases,Recoverd,Date\n
                Pakistan,20,100,80,20122019\n
                India,25,150,125,21112019\n
                USA,80,500,420,19012020\n
                Russia,120,300,180,11022020\n
                China,150,1000,850,31122019"""
                
csv_buffer = StringIO(covid_data)
df = pd.read_csv(csv_buffer)
df = pd.DataFrame(df)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d%m%Y')

df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year


# df = df.drop(columns=['Date'])
df
```

```python
moratality_rate = ((df['Deaths'] / df['Confirmed Cases'])*100).round(2).astype(str)+"%"
df["Mortality"] = moratality_rate
df
```

```python
updated_data = df.drop(columns='Date')
updated_data
```

```python
from io import StringIO
csv_buffer = StringIO()
updated_data.to_csv(csv_buffer,index=False)
csv_buffer = csv_buffer.getvalue()
print(csv_buffer)
```
