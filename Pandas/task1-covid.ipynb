###Aggragation

```python
grouped = df.groupby(by='Country/Region',as_index=False)['Confirmed'].sum()
print(grouped)
```

```python
grouped = df.groupby(by='Country/Region',as_index=False)['New cases'].sum
print(grouped)
```

```python
max_new_cases = df.groupby('Country/Region')['New cases'].max()

max_value = max_new_cases.max()

countries_with_max = max_new_cases[max_new_cases == max_value]
countries_with_max
```
